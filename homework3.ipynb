# 3.1图像旋转

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=cv2.imread('C:/Users/user/Desktop/Doraemon.jpg')
```

```python
rows,cols,channel=im.shape
```

```python
#参数：旋转中心 旋转度数 scale
M=cv2.getRotationMatrix2D((cols/2,rows/2),30,1)
#参数：原始图像 旋转参数 元素图像宽高
rotate_img=cv2.warpAffine(im,M,(cols,rows))
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(121), plt.imshow(im, 'gray'),plt.xlabel('Doraemon')
plt.subplot(122), plt.imshow(rotate_img, 'gray'),plt.xlabel('rotate Doraemon')
plt.tight_layout()
plt.show()
```

# 3.2 图像平移

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
im=cv2.imread('C:/Users/user/Desktop/Doraemon.jpg')
```

```python
#图像 下、上、左、右平移
M=np.float32([[1,0,0],[0,1,100]])
img1=cv2.warpAffine(im,M,(im.shape[1],im.shape[0]))

M=np.float32([[1,0,0],[0,1,-100]])
img2=cv2.warpAffine(im,M,(im.shape[1],im.shape[0]))

M=np.float32([[1,0,100],[0,1,0]])
img3=cv2.warpAffine(im,M,(im.shape[1],im.shape[0]))

M=np.float32([[1,0,-100],[0,1,0]])
img4=cv2.warpAffine(im,M,(im.shape[1],im.shape[0]))
```

```python
fig = plt.figure(figsize=(15, 16))
titles=['img1_up','img2_down','img3_left','img4_right']
images=[img1,img2,img3,img4]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
    plt.show
```
